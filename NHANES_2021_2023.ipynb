# Questions for Analysis

## Question 1

```python
## 3. Questions for Analysis

# Question 1: "Is there an association between marital status (married or not married) and education level (bachelor’s degree or higher vs. less than a bachelor’s degree)?"
  # Variables: DMDMARTZ (marital status) and DMDEDUC2 (education level). Recode as specified.

# Import necessary libraries
import pandas as pd
from scipy.stats import chi2_contingency
from IPython.display import display

# Extract relevant columns
demo_cleaned = demo [['SEQN', 'DMDMARTZ', 'DMDEDUC2']].copy()

# Recode Marital Status (DMDMARTZ)
demo_cleaned['Marital_Status'] = demo_cleaned['DMDMARTZ'].map({
    1: 'Married',          # Married/Living with partner
    2: 'Not Married',      # Widowed/Divorced/Separated
    3: 'Not Married',      # Never married
    4: 'Not Married'       # Separated
})

# Recode Education Level (DMDEDUC2)
demo_cleaned['Education_Level'] = demo_cleaned['DMDEDUC2'].map({
    5: 'Bachelor or Higher',       # College graduate or above
    1: 'Less than Bachelor',       # Less than 9th grade
    2: 'Less than Bachelor',       # 9-11th grade (Includes 12th grade with no diploma)
    3: 'Less than Bachelor',       # High school graduate/GED or equivalent
    4: 'Less than Bachelor'        # Some college or AA degree
})

# Drop rows with invalid or missing values
demo_cleaned = demo_cleaned.dropna(subset=['Marital_Status', 'Education_Level'])

# Cross-tabulate the recoded variables
association_table = pd.crosstab(demo_cleaned['Marital_Status'], demo_cleaned['Education_Level'])

# Perform chi-square test
chi2_stat, p_value, dof, expected_freq = chi2_contingency(association_table)

# Create a DataFrame for chi-square test results
chi_square_results_df = pd.DataFrame({
    "Metric": ["Chi2 Statistic", "P-value", "Degrees of Freedom"],
    "Value": [chi2_stat, p_value, dof]
})

# Create a DataFrame for expected frequencies
expected_freq_df = pd.DataFrame(
    expected_freq,
    index=association_table.index,
    columns=association_table.columns
)

# Display the results as formatted tables in Google Colab
print("Association Table:")
display(association_table)

print("\nChi-Square Test Results:")
display(chi_square_results_df)

print("\nExpected Frequencies:")
display(expected_freq_df)

print("\nStatistical Significance:")
print("A chi-square test was conducted, and the p-value was extremely small (p < 0.0001), indicating a statistically significant association between marital status and education level. ")
print("This means that the observed differences in education levels between married and not married individuals are unlikely to be due to random chance.")
print("\nConclusion:")
print("Married individuals are more likely to have higher education (bachelor’s or higher) compared to those who are not married.")
print("There are still many more married individuals with less than a bachelor’s degree overall.")
```

## Question 2

```python
# Question 2: "Is there a difference in the mean sedentary behavior time between those who are married and those who are not married?"
  # Variables: DMDMARTZ (marital status, recoded) and PAD680 (sedentary behavior time, cleaned).

# Import necessary libraries
import pandas as pd
from scipy.stats import ttest_ind
from IPython.display import display

# Extract and clean the marital status variable from DEMO_L dataset
demo_cleaned = demo [['SEQN', 'DMDMARTZ']].copy()

# Recode Marital Status (DMDMARTZ)
demo_cleaned['Marital_Status'] = demo_cleaned['DMDMARTZ'].map({
    1: 'Married',          # Married/Living with partner
    2: 'Not Married',      # Widowed/Divorced/Separated
    3: 'Not Married',      # Never married
    4: 'Not Married'       # Separated
})

# Drop rows with invalid or missing Marital_Status
demo_cleaned = demo_cleaned.dropna(subset=['Marital_Status'])

# Extract and clean the PAD680 variable from PAQ_L dataset
paq_cleaned = paq [['SEQN', 'PAD680']].copy()

# Clean the PAD680 variable (Sedentary Behavior Time)
# Remove values 7777, 9999, and null
paq_cleaned = paq_cleaned[(paq_cleaned['PAD680'] != 7777) & (paq_cleaned['PAD680'] != 9999)]
paq_cleaned = paq_cleaned.dropna(subset=['PAD680'])

# Merge marital status with sedentary behavior data
merged_data = pd.merge(demo_cleaned[['SEQN', 'Marital_Status']], paq_cleaned, on='SEQN')

# Group by marital status and calculate mean sedentary behavior time
sedentary_behavior_stats = merged_data.groupby('Marital_Status')['PAD680'].mean().reset_index()
sedentary_behavior_stats.rename(columns={'PAD680': 'Mean Sedentary Time (minutes)'}, inplace=True)

# Display the mean sedentary behavior times by marital status
print("Mean Sedentary Behavior Time by Marital Status:")
display(sedentary_behavior_stats)

# Perform a t-test to compare means between groups
# Split the data into groups based on marital status
married_sedentary = merged_data[merged_data['Marital_Status'] == 'Married']['PAD680']
not_married_sedentary = merged_data[merged_data['Marital_Status'] == 'Not Married']['PAD680']

# Perform the t-test (assuming unequal variances)
t_stat, p_value = ttest_ind(married_sedentary, not_married_sedentary, equal_var=False)

# Create a DataFrame for t-test results
t_test_results_df = pd.DataFrame({
    "Metric": ["T-Statistic", "P-value"],
    "Value": [t_stat, p_value]
})

# Display the t-test results
print("\nT-Test Results for Sedentary Behavior Time:")
display(t_test_results_df)

# Explaination
print("\nT-Test Results:")
print("The t-statistic is negative, indicating that the mean sedentary time is lower for married individuals compared to not married individuals.")
print("The p-value (0.000118) is less than 0.05, suggesting that the difference in mean sedentary behavior time between the two groups is statistically significant.")
```

## Question 3

```python
# Question 3: "How do age and marital status affect systolic blood pressure?"
  # Variables: RIDAGEYR (age), DMDMARTZ (marital status, recoded), and BPXOSY3 (systolic blood pressure).

# Import necessary libraries
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from IPython.display import display


# Extract and clean the relevant columns from DEMO_L dataset
age_bp_data = demo [['SEQN', 'RIDAGEYR', 'DMDMARTZ']].copy()

# Recode Marital Status (DMDMARTZ)
age_bp_data['Marital_Status'] = age_bp_data['DMDMARTZ'].map({
    1: 'Married',          # Married/Living with partner
    2: 'Not Married',      # Widowed/Divorced/Separated
    3: 'Not Married',      # Never married
    4: 'Not Married'       # Separated
})

# Drop rows with invalid or missing Marital_Status or Age
age_bp_data = age_bp_data.dropna(subset=['Marital_Status', 'RIDAGEYR'])

# Extract and clean the relevant columns from BPXO_L dataset
bp_cleaned = bp [['SEQN', 'BPXOSY3']].copy()

# Drop rows with missing or invalid systolic blood pressure values
bp_cleaned = bp_cleaned.dropna(subset=['BPXOSY3'])

# Merge datasets based on SEQN
merged_data_bp = pd.merge(age_bp_data[['SEQN', 'RIDAGEYR', 'Marital_Status']], bp_cleaned, on='SEQN')

# Group by marital status and calculate mean systolic blood pressure
bp_stats = merged_data_bp.groupby('Marital_Status')['BPXOSY3'].mean().reset_index()
bp_stats.rename(columns={'BPXOSY3': 'Mean Systolic Blood Pressure'}, inplace=True)

# Display the mean systolic blood pressure by marital status
print("Mean Systolic Blood Pressure by Marital Status:")
display(bp_stats)

# Analyze the effect of age and marital status on systolic blood pressure using a regression model
# Fit the model
model = ols("BPXOSY3 ~ RIDAGEYR + Marital_Status", data=merged_data_bp).fit()

# Extract regression results into a DataFrame
regression_results_df = pd.DataFrame({
    "Variable": model.params.index,
    "Coefficient": model.params.values,
    "Std Error": model.bse.values,
    "t-Statistic": model.tvalues.values,
    "P-value": model.pvalues.values
}).reset_index(drop=True)

# Display the regression results
print("\nRegression Results for Systolic Blood Pressure:")
display(regression_results_df)

# Explaination
print("\nMean Systolic Blood Pressure:")
print("Married individuals have a slightly lower mean systolic blood pressure compared to those not married.")
print("\nRegression Results:")
print("Age significantly increases systolic blood pressure (p < 0.001).")
print("Being not married increases systolic blood pressure slightly but significantly (p = 0.003).")
```

## Question 4

```python
# Question 4: "Is there a correlation between self-reported weight and minutes of sedentary behavior?"
  # Variables: WHD020 (self-reported weight, cleaned) and PAD680 (sedentary behavior time, cleaned).

import numpy as np
from scipy.stats import pearsonr

# Select relevant columns
paq_subset = paq [['SEQN', 'PAD680']]  # Sedentary behavior time
whq_subset = whq [['SEQN', 'WHD020']]  # Self-reported weight

# Merge datasets on SEQN
merged_data = pd.merge(paq_subset, whq_subset, on='SEQN', how='inner')

# Clean data:
# Remove rows where PAD680 or WHD020 have invalid values (7777, 9999, or null)
merged_data_cleaned = merged_data.dropna(subset=['PAD680', 'WHD020'])
merged_data_cleaned = merged_data_cleaned[
    ~merged_data_cleaned['PAD680'].isin([7777, 9999]) &
    ~merged_data_cleaned['WHD020'].isin([7777, 9999])
]

# Calculate the correlation between self-reported weight and sedentary behavior time
correlation, p_value = pearsonr(merged_data_cleaned['WHD020'], merged_data_cleaned['PAD680'])

# Display the results in a DataFrame
results = pd.DataFrame({
    'Metric': ['Correlation Coefficient', 'P-Value'],
    'Value': [correlation, p_value]
})

print("Correlation Results:")
display(results)

# Display basic statistics for both variables as context
summary_stats = merged_data_cleaned[['WHD020', 'PAD680']].describe().T.reset_index()
summary_stats.columns = ['Variable', 'Count', 'Mean', 'Std', 'Min', '25%', '50%', '75%', 'Max']

print("\nSummary Statistics:")
display(summary_stats)

# Explaination
print("\nConclusion:")
print("As self-reported weight increases, there is a slight tendency for sedentary behavior time to also increase. However, the relationship is not strong.")
```

## Question 5

```python
# Question 5: Does the level of education (DMDEDUC2) influence Vitamin D levels (LBDVD2LC) in adults aged 20 and older?"
  # Variables: DMDEDUC2 (level of education, cleaned) and LBDVD2LC (Vitamin D lab results, cleaned).

  # Hypothesis:
    # Null Hypothesis (H₀): There is no association between education level and Vitamin D levels.
    # Alternative Hypothesis (H₁): Education level significantly influences Vitamin D levels.

  # Test Chosen: Chi-Square Test
    # Reason:
      # DMDEDUC2 is a categorical variable representing education levels.
      # LBDVD2LC is a categorical variable with two levels: "At or Above Detection Limit" (0) and "Below Detection Limit" (1).
      # The chi-square test is suitable to determine whether the distribution of Vitamin D levels differs by education level.


# Import necessary libraries
import pandas as pd
from scipy.stats import chi2_contingency

# Recode education levels in DMDEDUC2
demo ['Education_Level'] = demo ['DMDEDUC2'].replace(
    {1: 'Less than Bachelor’s', 2: 'Less than Bachelor’s', 3: 'Less than Bachelor’s',
     4: 'Less than Bachelor’s', 5: 'Bachelor’s or Higher'}
)

# Recode Vitamin D Levels in LBDVD2LC
vid ['Vitamin_D_Level'] = vid ['LBDVD2LC'].replace({0: 'At or Above Detection Limit', 1: 'Below Detection Limit'})

# Merge the datasets on SEQN to combine education levels and Vitamin D levels
merged_data = pd.merge(demo [['SEQN', 'Education_Level']],
                       vid [['SEQN', 'Vitamin_D_Level']],
                       on='SEQN')

# Drop rows with missing data
cleaned_data = merged_data.dropna()

# Filter out unexpected values in Education_Level
cleaned_data_filtered = cleaned_data[cleaned_data['Education_Level'].isin(['Bachelor’s or Higher', 'Less than Bachelor’s'])]

# Create a contingency table
filtered_contingency_table = pd.crosstab(cleaned_data_filtered['Education_Level'], cleaned_data_filtered['Vitamin_D_Level'])

# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(filtered_contingency_table)

# Display final results in a dataframe
results_df = pd.DataFrame({
    "Chi-Square Statistic": [chi2],
    "Degrees of Freedom": [dof],
    "P-Value": [p]
})

# Display the contingency table as a dataframe
contingency_table_df = filtered_contingency_table.reset_index()

# To ensure display in Google Colab
from IPython.display import display

# Show both dataframes
print("Contingency Table:")
display(contingency_table_df)

print("\nChi-Square Test Results:")
display(results_df)

print("\nSummary of Findings:")
print("The P-value is significantly below the threshold of 0.05, suggesting a statistically significant relationship between education level and Vitamin D levels.")
print("Individuals with a 'Bachelor's or Higher' degree are more likely to have 'At or Above Detection Limit' Vitamin D levels compared to those with 'Less than Bachelor's'.")
```
